##Why

 ---

##How
- Elbow Method

**Why**Used to help determine a reasonable value for K (number of clusters).

  - We calculated WCSS (Within-Cluster Sum of Squares) for different K values.

  - Observed how WCSS decreases as K increases.

  - The ‚Äúelbow point‚Äù indicates where adding more clusters stops giving significant improvement.

  - Based on this curve, we identified the most stable and meaningful K values to test in our model.

‚Ä¢ Silhouette Score

**Why:** Used to evaluate the quality of the clusters created by K-Means.

  - Measures how well each data point fits within its assigned cluster (cohesion) and how far it is from other clusters (separation).

  - Score ranges from -1 to 1; higher scores indicate better-defined clusters.

  - We computed the Silhouette Score for each K value we tested.

  - This allowed us to compare which K provides the best clustering structure and avoid poor or overlapping clusters.

‚Ä¢ PCA (Principal Component Analysis)

**Why:** Applied only for visualization, not for training the model.

  - Reduces the dataset to two principal components so that clusters can be plotted clearly on a 2D graph.

  - Helps visually confirm whether clusters are separated or overlapping.

  - Used to present scatter plots showing how students group based on their social media usage patterns.

‚Ä¢ Final K-Means Implementation Summary

 - Removed the target label since clustering is unsupervised.

 - Tested multiple K values (e.g., 2, 3, 4).

 - For each K:

   - Ran K-Means

   - Calculated WCSS (for Elbow)

   - Calculated Silhouette Score

   - isualized the clusters using PCA

Selected the most meaningful K based on both numerical evaluation and visual separation.

- Both criteria produced very similar performance across all train‚Äìtest partitions (90‚Äì10, 80‚Äì20, 70‚Äì30).

- The results indicate that both splitting measures behave almost identically on our dataset, suggesting that the dataset does not strongly favor one impurity measure over the other.

- Minor accuracy differences are expected because Entropy tends to penalize mixed nodes more, while Gini is computationally faster‚Äîbut in our case, this did not impact the final performance.

Overall, Entropy performed marginally better, but both criteria are equally reliable for this dataset..
